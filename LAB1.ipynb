```python
import re
import random

# Saludos
saludo = re.compile(r"\b(hola|qué tal|buenos días|buenas tardes|buen dia|que onda|hey|hello)\b", re.IGNORECASE)
# Eliza
nombre = re.compile(r"\b(eliza)\b", re.IGNORECASE)
# Insultos
insulto = re.compile(r"\b(tonta|estúpida|idiota|puto|puta|maricon|cerote|cerota|tarado)\b", re.IGNORECASE)

# Patrones
PATRONES = [
    (re.compile(r"me siento (.*)", re.IGNORECASE),
     ["¿Por qué te sientes {0}?", "¿Desde cuándo te sientes {0}?", "¡Qué bien que te sientas {0}!", "Lamento mucho que te sientas {0}"]),
    (re.compile(r"estoy (.*)", re.IGNORECASE),
     ["¿Qué te hace estar {0}?", "¿Cómo influye estar {0} en tu día?", "¿Te gustaría hablar más sobre estar {0}?"]),
    (re.compile(r"mi (.*)", re.IGNORECASE),
     ["Cuéntame más sobre tu {0}.", "¿Y qué pasa con tu {0}?", "¿Cómo te afecta tu {0}?"]),
    (re.compile(r"(.*) trabajo (.*)", re.IGNORECASE),
     ["¿Te gusta trabajar {1}?", "¿Desde cuándo trabajas en {1}?", "¿Cómo te hace sentir trabajar {1}?"]),
    (re.compile(r"(.*)\?", re.IGNORECASE),
     ["Esa es una buena pregunta.", "¿Por qué te interesa saber eso?", "¿Qué esperas encontrar al preguntar eso?"]),
    (re.compile(r"quiero (.*)", re.IGNORECASE),
     ["¿Por qué quieres {0}?", "¿Qué te motiva a querer {0}?"]),
    (re.compile(r"necesito (.*)", re.IGNORECASE),
     ["¿Por qué sientes que necesitas {0}?", "¿Cómo crees que podrías conseguir {0}?"]),
    (re.compile(r"me gusta (.*)", re.IGNORECASE),
     ["¿Qué es lo que más te gusta de {0}?", "¿Por qué te gusta {0}?", "¿Cómo te hace sentir {0}?"]),
]


def es_saludo(texto: str) -> bool:
    return bool(saludo.search(texto))

def es_nombre(texto: str) -> bool:
    return bool(nombre.search(texto))

def es_insulto(texto: str) -> bool:
    return bool(insulto.search(texto))

def match_patron(texto: str):
    for patron, respuestas in PATRONES:
        m = patron.search(texto)
        if m:
            grupos = m.groups()
            plantilla = random.choice(respuestas)
            return plantilla.format(*grupos), True
    return None, False

# Conversación inicial
conversacion_iniciada = False
def responder(mensaje: str) -> str:
    global conversacion_iniciada

    # Saludo
    if not conversacion_iniciada:
        if es_saludo(mensaje):
            conversacion_iniciada = True
            return random.choice(["¡Hola!", "¿Qué tal?", "Que onda!"])
        else:
            return "Es importante iniciar una conversación con un saludo."

    # Eliza
    if es_nombre(mensaje):
        return "Hola, ¿cómo estás?"

    # Insulto
    if es_insulto(mensaje):
        return "No me trates así."

    # Patrones
    respuesta, hay_patron = match_patron(mensaje)
    if hay_patron:
        return respuesta

    # No encuentra nada
    return "No puedo comprender tu comentario."

```
